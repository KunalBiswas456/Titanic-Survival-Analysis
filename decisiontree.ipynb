```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
import matplotlib.pyplot as plt
import seaborn as sns
print(os.listdir("../input"))
```

```python
df=pd.read_csv("../input/train.csv")
test=pd.read_csv("../input/test.csv")
```

```python
print("Number of Missing Values:")
print(df.isna().sum())
df=df.dropna(axis=1)
print("\nData Cleaned:")
print(df.isna().sum())
```

```python
#Remove Name, Ticket Number, Cabin Number, SibSp, Parch
df=df.drop(["Name","Ticket", "SibSp", "Parch"],axis=1)
```

```python
#More passengers are from class Three. find the gender ratio among the classes
sns.factorplot('Pclass',data=df,hue='Sex',kind='count')
```

```python
sns.factorplot('Survived',data=df,kind='count',hue='Pclass')
```

```python
#Convert Pclass(1,2,3) to Pclass_1, Pclass_2, Pclass_3
df=pd.get_dummies(df,columns=["Pclass"])
```

```python
#Transform string values "Male"->0 and "Female"->1
df=df.replace("male",0)
df=df.replace("female",1)
```

```python
df.head()
```

```python
#Divide the train dataset into two parts
#70% for model
#30% for testing
from sklearn.model_selection import train_test_split
xtrain,xtest=train_test_split(df,test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf_dtc=DecisionTreeClassifier()
```

```python
#Separate Prediction Label from testing Set
f=xtrain["Survived"]
xtrain=xtrain.drop("Survived",axis=1)
```

```python
#Train the DecisionTree Model
clf_dtc.fit(xtrain,f)
```

```python
#Separate the "Survived" Label
ft=xtest["Survived"]
xtest=xtest.drop("Survived",axis=1)
```

```python
clf_dtc.score(xtest,ft)
```
